```python
import numpy as np
import pandas as pd
df=pd.read_csv("C:\\Users\\Charan\\Downloads\\BP_data.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
a=df.drop(['Pregnancy','Blood_Pressure_Abnormality'],axis=1)

df.fillna(df.mean(), inplace=True)
a.fillna(a.mean(), inplace=True)

a
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

# separate independent & dependent variables
X = a.iloc[:,0:12]  #independent columns
y = df.iloc[:,1]    #target column i.e price range

# apply SelectKBest class to extract top 10 best features
bestfeatures = SelectKBest(score_func=chi2, k=10)
fit = bestfeatures.fit(X,y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)

#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']  #naming the dataframe columns
print(featureScores.nlargest(11,'Score'))  #print 10 best features
```

```python
featureScores = featureScores.sort_values(by='Score', ascending=False)
featureScores
```

```python
plt.figure(figsize=(25,10))
sns.barplot(x='Specs', y='Score', data=featureScores, palette = "GnBu_d")
plt.box(False)
plt.title('Feature importance', fontsize=16)
plt.xlabel('\n Features', fontsize=14)
plt.ylabel('Importance \n', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
y=pd.DataFrame(df)
y
```

```python
x=np.array(y)
x
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(x)
y_Kmeans = kmeans.predict(x)
print(y_Kmeans)
```

```python
print(kmeans.cluster_centers_)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x[:, 9], x[:, 10], c=y_Kmeans)
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 9], centers[:, 10],c='blue')
```
